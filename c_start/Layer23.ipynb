```python
import numpy as np

w1 =(-125,0)
w2 =(125,45)
w3 =(125,-45)

def eqn_line(p1,p2):
    x1,z1 =p1
    x2,z2 =p2
    if x1==x2:
        a=1
        b=0
        c=-x1
    else:
        a=z2-z1
        b=x1-x2
        c=(x1*(z1-z2))+(z1*(x2-x1))
    return a,b,c

def distance(a,b,c,x0,y0):
    r = abs((a*x0)+(b*y0)+c)/(np.sqrt((a**2)+(b**2)))
    return r

# changing wedge 1, axis joining W2-W3
a,b,c = eqn_line(w2,w3)
print(f"The coefficients of the Eqn of line joining W2 & W3 is:\n"
      f"a,b,c = {a,b,c}")
r = distance(a,b,c,-125,0)
print(f"perpendicular distance from W1 joing axis W2 & W3 is \n"
      f"d1 =:{r}")
r = distance(a,b,c,0,0)
print(f"perpendicular distance from Magnet center joing axis W2 & W3 is \n"
      f"m1 =:{r}")

# changing wedge 2, axis joining W1-W3
a,b,c = eqn_line(w1,w3)
print(f"The coefficients of the Eqn of line joining W1 & W3 is:\n"
      f"a,b,c = {a,b,c}")
r = distance(a,b,c,125,45)
print(f"perpendicular distance from W2 joining axis W1 & W3 is  \n"
      f"d2 =:{r}")
r = distance(a,b,c,0,0)
print(f"perpendicular distance from Magnet center joing axis W1 & W3 is \n"
      f"m2 =:{r}")


# changing wedge 3, axis joining W1-W2    
a,b,c = eqn_line(w1,w2)
print(f"The coefficients of the Eqn of line joining W1 & W2 is: \n"
      f"a,b,c = {a,b,c}")
r = distance(a,b,c,125,-45)
print(f"perpendicular distance from W3 joing axis W1 & W2 is \n"
      f"d3 =:{r}")
r = distance(a,b,c,0,0)
print(f"perpendicular distance from Magnet center joing axis W1 & W2 is \n"
      f"m3 =:{r}")

# Calculating the angle
def change_lay3(n1,n2,n3):
    d1 = 250
    d2 = 88.57649915763257
    d3 = 88.57649915763257
    h  = 0.1
    m1 = 125.0
    m2 = 22.144124789408142
    m3 = 22.144124789408142 
    alpha = (n1*h)/d1
    beta  = (n2*h)/d2
    gamma = (n3*h)/d3
    y  = (m3*gamma)+(m2*beta)+(m1*alpha)
    return alpha, beta,gamma,y
 
############################################################################################
#Giving the number of rotation
alpha,beta,gamma,y = change_lay3(1,0,0)

print(f"angle corresponding to wedge 1 (Alpha): {alpha}\n"
      f"angle corresponding to wedge 2 (Beta): {beta}\n"
      f"angle corresponding to wedge 3 (Gamma): {gamma}\n"
      f"distance along y\n y={y}"
      )

```
